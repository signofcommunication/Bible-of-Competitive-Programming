```python
import pandas as pd
import random
from datetime import datetime
import matplotlib.pyplot as plt

# === 1. Load Bank Soal ===
bank = pd.read_csv("bank-soal.csv")

# === 2. Ambil 20 Soal Secara Random ===
soal_main = bank.sample(20).reset_index(drop=True)

jawaban_user = []
hasil = []

print("=== MULAI GAME SCIENCE QUIZ (ESSAY) ===\n")

# === 3. Loop Menanyakan Soal ===
for i, row in soal_main.iterrows():
    print(f"Soal {i+1}: {row['soal']}")
    user_answer = input("Jawaban Anda: ").strip()

    # Penilaian sederhana (jawaban harus mengandung kata kunci)
    benar = user_answer.lower() in row['jawaban'].lower()

    jawaban_user.append(user_answer)
    hasil.append({
        "soal": row["soal"],
        "jawaban_user": user_answer,
        "jawaban_benar": row["jawaban"],
        "kategori": row["kategori"],
        "poin": row["poin"] if benar else 0,
        "status": "Benar" if benar else "Salah"
    })

    print(f"Status: {'✔️ Benar' if benar else '❌ Salah'}\n")

# === 4. Statistik ===
df_hasil = pd.DataFrame(hasil)

total_benar = df_hasil[df_hasil["status"]=="Benar"].shape[0]
total_salah = df_hasil[df_hasil["status"]=="Salah"].shape[0]
total_poin = df_hasil["poin"].sum()

print("\n=== HASIL AKHIR ===")
print(f"Total Benar : {total_benar}")
print(f"Total Salah : {total_salah}")
print(f"Total Poin  : {total_poin}")

# tampilkan tabel hasil
display(df_hasil)


# =====================
#     GRAFIK STATISTIK
# =====================

# --- Bar Chart ---
plt.figure(figsize=(8,5))
plt.bar(["Benar", "Salah", "Total Poin"],
        [total_benar, total_salah, total_poin])
plt.title("Statistik Jawaban")
plt.ylabel("Jumlah")
plt.show()

# --- Pie Chart ---
plt.figure(figsize=(6,6))
plt.pie(
    [total_benar, total_salah],
    labels=["Benar", "Salah"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Persentase Jawaban")
plt.show()


# =====================
#     SIMPAN KE CSV
# =====================

timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
filename = f"hasil_quiz_{timestamp}.csv"

df_hasil.to_csv(filename, index=False)

print(f"\nHasil permainan sudah disimpan ke file: {filename}")
```
